```python
def asal_mi(sayi):
    # Numbers less than 2 are not prime
    if sayi < 2:
        return False
    # 2 is the only even prime number
    elif sayi == 2:
        return True
    else:
        # Check for divisibility from 2 up to the square root of sayi
        # If a number is not prime, it must have a divisor less than or equal to its square root.
        for i in range(2, int(sayi**0.5) + 1):
            if sayi % i == 0:
                return False  # Found a divisor, so it's not prime
        return True  # If loop finishes without returning False, it's prime

print(asal_mi(9))
```
