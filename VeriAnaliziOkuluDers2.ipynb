```python
 # bir sayı dizisinde tek ve çift rakamları belirleyen kod.

import random

liste = [random.randint(0, 99) for _ in range(10)]
çift = []
tek = []

for sayi in liste:
  if sayi % 2 == 0:
    çift.append(sayi)
  else:
    tek.append(sayi)
print ("çift sayılar: ", çift )
print ("tek sayılar: ", tek)
```
