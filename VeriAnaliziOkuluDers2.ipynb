
## Fonksiyonun Temel Bileşenleri
### Fonksiyon Başlığı (Function Header)
Fonksiyon, def anahtar kelimesi ile başlar, ardından fonksiyonun adı ve parantez içinde aldığı parametreler (argümanlar) gelir.
Örneğin, önceki örnekte fonksiyonun adı greeting, parametresi ise name idi.

### Fonksiyon Dokümantasyonu (Docstring)
Fonksiyonun ne yaptığını açıklamak için kullanılır.
Genellikle """üç tırnak işareti""" ile yazılır.

### Fonksiyon Gövdesi (Function Body)
Fonksiyonun yaptığı işlemler buraya yazılır.
Fonksiyon çağrıldığında bu kısım çalıştırılır.

```python
import random # random modülünü içe aktarıyoruz

# listedeki tüm sayıların çarpımını veren fonksiyon
def carpi(): # Fonksiyon artık dışarıdan bir 'liste' argümanı almıyor
  liste = [random.randint(0, 10) for _ in range(5)] # 0-10 arası 5 rastgele sayıdan oluşan liste oluştur
  carpim = 1 # Çarpımı tutacak değişkeni 1 ile başlatıyoruz
  for sayi in liste: # Listenin elemanları üzerinde döngü yap
    carpim *= sayi # Her elemanı çarpıma ekle
  print("Oluşturulan liste:", liste) # Oluşturulan listeyi yazdır
  return carpim # Çarpım sonucunu döndür

# Fonksiyonu çağır ve sonucunu yazdır
print("Listenin elemanlarının çarpımı:", carpi())
```
