### Sözlükler- dictionaries

Sözlükler sıralı olarak saklama yapmaz, indeksle ulaşılamaz. sözlükler key-value ikilileri şeklinde tutar. { parantez ile gösterilir. : ile tanımlanır.

```python
Movie = {
    "name": "The Shawshank Redemption",
    "year": 1994,
    "director": "Frank Darabont"
}
```

```python
print(Movie.keys())
```

```python
print(Movie.values())
print(Movie.items())
```

```python
Movie.get("name")
```

```python
Movie['year'] = 2026
```

```python
print(Movie)
```

```python
Movie.update({"genre": "dram"})
Movie
```
